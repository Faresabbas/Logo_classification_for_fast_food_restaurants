```python
classes = {'McDonalds': 0, 'Starbucks': 1, 'Other': 5, 'Burger King': 2, 'KFC': 3, 'Subway': 4}

def get_classes(n):
    for x, y in classes.items():
        if n == y:
            return x
     
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)

y_train = np.array(y_train)
y_test = np.array(y_test)

print(f'X_train shape  is {X_train.shape}')
print(f'X_test shape  is {X_test.shape}')
print(f'y_train shape  is {y_train.shape}')
print(f'y_test shape  is {y_test.shape}')
```

```python
import keras
KerasModel = keras.models.Sequential([
        keras.layers.Conv2D(512,kernel_size=(3,3),activation='relu',input_shape=(64,64,3)),
        keras.layers.Conv2D(512,kernel_size=(3,3),activation='relu'),
        keras.layers.MaxPool2D(4,4),
        keras.layers.Conv2D(256,kernel_size=(3,3),activation='relu'),    
        keras.layers.Conv2D(128,kernel_size=(3,3),activation='relu'),    
        keras.layers.Conv2D(64,kernel_size=(3,3),activation='relu'),
        keras.layers.MaxPool2D(4,4),
        keras.layers.Flatten() ,    
        keras.layers.Dense(128,activation='relu') ,    
        keras.layers.Dense(64,activation='relu') ,    
        keras.layers.Dense(64,activation='relu') ,        
        keras.layers.Dropout(rate=0.5) ,            
        keras.layers.Dense(6,activation='softmax') ,    
        ])
```

```python
KerasModel.compile(optimizer ='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
print('Model Details are : ')
print(KerasModel.summary())
```

```python
epochs = 25
ThisModel = KerasModel.fit(X_train, y_train, epochs=epochs,batch_size=64,verbose=1)
```

```python
ModelLoss, ModelAccuracy = KerasModel.evaluate(X_test, y_test)

print('Test Loss is {}'.format(ModelLoss))
print('Test Accuracy is {}'.format(ModelAccuracy ))
```

```python
import tensorflow as tf
import numpy as np
from IPython.display import display, HTML
import ipywidgets as widgets
from PIL import Image

# Define the path to your pre-trained model (replace with your model's path)
model_path = '/kaggle/working/my_model.h5'  # Example path, replace with your model's path
class_names =['McDonalds', 'Starbucks', 'Other', 'Burger King', 'KFC', 'Subway']  # Replace with your own class names

# Load the pre-trained model
model = tf.keras.models.load_model(model_path)

# Function to preprocess and make predictions
def predict_uploaded_image(change):
    uploaded_image = change['new'][0]
    
    # Load and preprocess the uploaded image
    img = Image.open(uploaded_image)
    img = img.resize((64, 64))  # Resize to match model input size
    img = np.array(img) / 255.0  # Normalize pixel values
    
    # Make predictions
    predictions = model.predict(np.expand_dims(img, axis=0))
    predicted_class_index = np.argmax(predictions)
    predicted_class = class_names[predicted_class_index]
    
    # Display the prediction
    print(f"Predicted Class: {predicted_class}")
    print("Confidence Scores:")
    for i, class_name in enumerate(class_names):
        confidence = predictions[0][i]
        print(f"{class_name}: {confidence:.2f}")

# Create an upload button for image upload
upload_btn = widgets.FileUpload(
    accept='image/*',
    multiple=False
)

# Attach the event handler to the upload button
upload_btn.observe(predict_uploaded_image, names='value')

# Display the upload button
display(upload_btn)
```
